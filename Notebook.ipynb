### Cleaning Data

```python
datasets = [Aotizhongxin_df, Changping_df, Dingling_df,
            Dongsi_df, Guanyuan_df, Gucheng_df, Huairou_df,
            Nongzhanguan_df, Shunyi_df, Tiantan_df, Wanliu_df, Wanshouxing_df]

for df in datasets:
  mode_wd = df['wd'].mode()[0]
  df['wd'].fillna(mode_wd, inplace=True)
```

```python
datasets = [Aotizhongxin_df, Changping_df, Dingling_df,
            Dongsi_df, Guanyuan_df, Gucheng_df, Huairou_df,
            Nongzhanguan_df, Shunyi_df, Tiantan_df, Wanliu_df, Wanshouxing_df]

null_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

for df in datasets:
  for i in null_columns:
    mean_value = df[i].mean()
    df[i].fillna(mean_value, inplace=True)
```

## Visualisasi Data

### Pertanyaan 1 - Jawaban

1. Bagaimana perkembangan persebaran polutan dari tahun ke tahun?

  Kenaikan polutan dapat dilihat terjadi secara fluktatif sejak 2013 hingga 2017. Dengan indikator utama PM2.5 sebagai acuan polutan halus yang berbahaya, didapatkan peningkatan besar-besaran terjadi pada tahun 2016 - 2017 yaitu sebesar **28.31%** berada pada angka konsentrasi 92.5 dari 72.5 (dapat dilihat pada line chart di bawah ini).

  Persentase tersebut dapat dihitung dari rata - rata nilai yang didapatkan dalam masing - masing tahun tersebut dengan rumus:

  $$
  \text{Persentase Kenaikan} = \left( \frac{\text{jumlah nilai rata - rata akhir} - \text{jumlah nilai rata - rata awal}}{\text{jumlah nilai rata - rata awal}} \right) \times 100
  $$

  Pertimbangan penggunaan variable PM2.5 dalam pembacaan Air Quality Index karena PM2.5 tersedia secara luas dan dianggap sebagai polutan udara paling berbahaya yang berdampak pada kesehatan manusia. Dengan susunan partikular materi seperti debu, kotoran, asap, dan tetesan cair.

  Sumber:
  - [Sumber 1](https://www.iqair.com/id/newsroom/what-is-aqi?srsltid=AfmBOooA0FR-u3MBTwfnjh5oEiWCeG8D22jeTI-i_y1IcN1-X8gdKs-P)
  - [Sumber 2](https://www.iqair.com/id/newsroom/pm2-5?srsltid=AfmBOorC3XaGeaUd8WTCz-DxBa-TFCnR087r7Qfvn5b2_ielkwtepM42)


```python
yearly_avg = all_data_air_quality.groupby(['year'])[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='PM2.5', data=yearly_avg, marker='o')
plt.title('Tren PM2.5 Tahunan')
plt.xlabel('Tahun')
plt.ylabel('Konsentrasi PM2.5')
plt.show()

for i in range(1, len(yearly_avg)):
    nilai_awal = yearly_avg['PM2.5'][i-1]
    nilai_akhir = yearly_avg['PM2.5'][i]
    persentase_kenaikan = ((nilai_akhir - nilai_awal) / nilai_awal) * 100
    print(f"Kenaikan dari {yearly_avg['year'][i-1]} ke {yearly_avg['year'][i]}: {persentase_kenaikan:.2f}%")
```

```python
for i in range(1, len(yearly_avg)):
    nilai_awal = yearly_avg['PM2.5'][i-1]
    nilai_akhir = yearly_avg['PM2.5'][i]
    persentase_kenaikan = ((nilai_akhir - nilai_awal) / nilai_awal) * 100
    print(f"Kenaikan dari {yearly_avg['year'][i-1]} ke {yearly_avg['year'][i]}: {persentase_kenaikan:.2f}%")
```

### Pertanyaan 2 - Jawaban

2. Bagaimana persebaran polutan dapat semakin meningkat?

  Persebaran polutan dapat meningkat karena tingginya temperature lingkungan yang diikuti dengan tingkat kelembapan udara. Dapat dilihat pada variable TEMP (Suhu Udara / Temperature) dan DEWP (kelembapan udara / DEW Point) yang mencapai korelasi **0.82.**

  Hal ini dapat mengindikasikan bahwa suhu udara yang tinggi dapat mempercepat reaksi kimia dalam atmosfer untuk menghasilkan partikel PM2.5, tingginya peningkatan suhu ini juga dapat dipengaruhi dari aktivitas sehari - hari seperti penggunaan kendaraan bermotor dan industri. Hal ini juga diikuti jika kelembapan tinggi yang juga mendukung terjadinya pengendapan partikel PM2.5 lebih cepat karena uap air.

  Hal ini juga berbanding lurus dengan diagram pada rata - rata indeks polutan PM2.5 dalam 24 jam. Peningkatan dalam persebaran partikel polutan mencapai titik tertinggi saat aktivitas masyarakat berjalan diikuti dengan kelembapan suhu yang meningkat. Dengan kata lain, peningkatan terjadi mulai dari jam 13:00 hingga jam 23:00.
